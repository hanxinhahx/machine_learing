```python
from matplotlib.image import imread
from sklearn.cluster import KMeans
import numpy as np
```

```python
image=imread('./data/flower.jpg')
image.shape
```

```python
X=image.reshape(-1,3) # 像素点展平，仍保留3个通道
X.shape
```

```python
kmeans=KMeans(n_clusters=4,random_state=42).fit(X) # n_clusters表示要分成的聚簇数量
kmeans.cluster_centers_ # 得到中心点
```

```python
segmented_img=kmeans.cluster_centers_[kmeans.labels_].reshape(image.shape) # 得到每一个中心点所对应聚簇中的其他点，并还原成原始大小
segmented_img
```

### 对比实验

```python
segmented_imgs=[]
n_colors=[8,6,4,2] # 指定不同的聚簇个数
for n_cluster in n_colors:
    kmeans=KMeans(n_clusters=n_cluster,random_state=42).fit(X)
    segmented_img=kmeans.cluster_centers_[kmeans.labels_]
    segmented_imgs.append(segmented_img.reshape(image.shape))
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(8,8))
for idx,n_clusters in enumerate(n_colors):
    plt.subplot(221+idx)
    plt.imshow(segmented_imgs[idx].astype(np.uint8))
    plt.title('{} colors'.format(n_clusters))
plt.figure(figsize=(5,5))
plt.imshow(image)
plt.title('Original Image')
plt.show()
```
